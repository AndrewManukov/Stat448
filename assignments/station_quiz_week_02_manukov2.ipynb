````sas
filename loan1 url  "https://uofi.box.com/shared/static/vi37omgitiaa2yyplrom779qvwk1g14x.csv";

data sba;
 infile loan1 dsd firstobs=2 missover;
 input LoanNr_ChkDgt Name $ City $ State $ Zip
      Bank $ BankState $ NAICS ApprovalDate :date9.
      ApprovalFY Term NoEmp NewExist
      CreateJob RetainedJob FranchiseCode Urban_Rural
      RevLineCr $ LowDoc $ ChgOffDate :date9.
      DisbursementDate :date9. DisbursementGross :dollar15.2
      BalanceGross :dollar15.2 MIS_Status $
      ChgOffPrinGr :dollar15.2 GrAppv :dollar15.2 SBA_Appv :dollar15.2;
format ApprovalDate mmddyy10. ChgOffDate mmddyy10. DisbursementDate mmddyy10.
       DisbursementGross dollar15.2 BalanceGross dollar15.2 ChgOffPrinGr dollar15.2
       GrAppv dollar15.2 SBA_Appv dollar15.2;
run;

proc contents data=sba;
run;
proc print data=sba (obs=10);
run;

data sba_len;
length Name $ 20 City $ 20 Bank $ 20;
set sba;
run;

/*With SAS, create a new subset of the SBA Business Loans Data that contains the following 8 variables:
 Name, City, State, ApprovalFY (approval year), NoEmp (number of employees), MIS_Status,
 GrAppv (gross approved amount), and SBA_Appv (SBA approved amount).
Now, show only the first 10 rows of the new subset with the columns in this order of appearance
sorted by ApprovalFY from oldest to newest order. (SAS, Data Wrangling)*/

data sba_sub;
set sba_len;
keep Name City State ApprovalFY NoEmp MIS_Status GrAppv SBA_Appv;
run;

proc sort data=sba_sub;
by descending ApprovalFY;
run;

proc print data = sba_sub (obs=10);
run;
````

````sas
*Part 8

data sba_3;
 set sba_2;
 Total_Employees=NoEmp;
 Approval_Year=ApprovalFY;
 Loan_Status=MIS_Status;
 Gross_Approved_Amount=GrAppv;
 SBA_Approved_Amount=SBA_Appv;
 SBA_Guranteed_Proportion = SBA_Approved_Amount/Gross_Approved_Amount;
 if Total_Employees= 1 then Business_Size="Small";
 if Total_Employees <= 20 and  Total_Employees >=  2 then Business_Size="Medium";
 if Total_Employees >  20 then Business_Size="Large";
 drop NoEmp ApprovalFY MIS_Status GrAppv SBA_Appv;
run;
proc sort data=sba_3;
    by Approval_Year;
run;
proc print data=sba_3 (obs=10);
    where city="Dover" and state="DE";
run;
````

````sas
*Part 9
data basicstats;
    set sba_3;
    if Name="" then delete;
    if City="" then delete;
    if State="" then delete;
    if Total_Employees=. then delete;
    if Approval_Year=. then delete;
    if Loan_Status="" then delete;
    if Gross_Approved_Amount=. then delete;
    if SBA_Approved_Amount=. then delete;
    if 	SBA_Guranteed_Proportion=. then delete;
    if 	Business_Size="" then delete;
run;
proc univariate data=basicstats (obs=15);
class Total_Employees  Approval_Year;
var Gross_Approved_Amount;
run;
````

````sas
*Part 10
data props;
    set sba_3;
    if Name="" then delete;
    if City="" then delete;
    if State="" then delete;
    if Total_Employees=. then delete;
    if Approval_Year=. then delete;
    if Loan_Status="" then delete;
    if Gross_Approved_Amount=. then delete;
    if SBA_Approved_Amount=. then delete;
    if 	SBA_Guranteed_Proportion=. then delete;
    if 	Business_Size="" then delete;
run;
proc means data=props (obs=10) mean;
class Loan_Status  State;
var SBA_Guranteed_Proportion;
run;
````
